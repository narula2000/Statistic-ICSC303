### Linear Regression

```r
## Here's the data from the video
mouse.data <- data.frame(
  weight=c(16, 11, 15, 8, 12, 10, 13, 14),
  size=c(58, 40, 55, 35, 43, 41, 52, 49))

mouse.data # print the data to the screen in a nice format

## plot a x/y scatter plot with the data
plot(mouse.data$weight, mouse.data$size)
```

```r
## create a "linear model" - that is, do the regression
mouse.regression <- lm(size ~ weight, data=mouse.data)

## generate a summary of the regression
summary(mouse.regression)

## add the regression line to our x/y scatter plot
plot(mouse.data$weight, mouse.data$size) + abline(mouse.regression, col="blue")

## Pearson Correlation
cat("Pearson Correlation:", cor(mouse.data$weight, mouse.data$size, method = "pearson"))
```

### AVONA Test

```r
## Combine all group
combineG <- data.frame(
    group1=c(121.3, 111.9, 110.1, 105.4, 101.6), 
    group2=c(99.5, 113.2, 108.9, 109.1, 100.4), 
    group3=c(104.2, 109.7, 102.3, 111.2, 106.6))

summary(combineG)

## Do the AVONA test
answer <- aov(values ~ ind,data = stack(combineG))
summary(answer)
```

### T-Test

```r
data <- c(12.5, 13.0, 11.9, 10.2, 13.1, 13.6, 13.8, 14.0)

## Can adjust mu value or leave it to defualt
t.test(data, mu=13.5, conf.level=0.95)

## Critical Value qt(alpha, df)
cat("Critical Value(s) =", qt(0.05, df=7))
```

### Confidence Interval T-Student Distribution 
We usually use qt() rather than qnorm() cause the data size is small.
$$ 
   Mean \pm T \cdot StandardError
$$

$$
    t_{\frac{\alpha}{2}} = qt(1 - \frac{\alpha}{2}, df)
$$

```r
data <- c(5.6,5.4,5.2,5.4,5.7,5.5,5.6,5.4,5.3,5.5,5.5,5.1,5.6,5.4,5.4)

alpha <- 0.05
len <- length(data)
df <- len - 1
mean <- mean(data)

tHalf <- qt(1 - alpha / 2, df=df)

standE <- sd(data) / sqrt(len)

cat("Confidence Interval:", mean + c(-1, 1) * tHalf * standE)
```

### Standard Error
$$
    StandardError = \sum{\frac{StandardDiv}{\sqrt{n}}}
$$

### Tcal or Tval
$$
    Tcal = \frac{\Delta mean}{StandardError}
$$

Test with T-half so check $ |Tcal| >  t_{\frac{\alpha}{2}} $

```r
data = data.frame(
  "Statistics"=c("Sample Size", "Sample Mean", "Sample SD"),
  "Southern Cal Edison"=c(9,3.2,0.5),
  "Southern Cal Gas"=c(12,2.8,0.7)) 

data

alpha <- 0.10

SE <- sqrt(0.5^2 / 9 + 0.7^2 / 12)
tCal <- (3.2 - 2.8) / SE

print(paste0("tCal =", tCal, ", tHalf =", qt(1 - (alpha / 2), min(9 -1, 12 - 1))))
```

tCal not greater than tHalf, Can't reject Null hypo

### Chi-Square Test

```r
data = data.frame(
    "Position"=c("Faculty","Staff","Students"), 
    "Walk"=c(19 ,14,27), 
    "Bike"=c(28,20,49), 
    "Automobile"=c(75,63,88), 
    "Other"=c(45,70,67))
data

mtx = matrix(c(19,28,75,45,14,20,63,70,27,49,88,67), nrow=3, byrow=T)
mtx

chisq.test(mtx)
```

Check with the table using df

```r
qchisq(0.95,6)
```

$X^2$ is greater than the table value hence reject Null hypo, We can also say p-val greater than alpha hence reject.
